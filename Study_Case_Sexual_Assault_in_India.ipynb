```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/Detailed Cases (Registered) sexual Assault 2001-2008.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={
    "Incest (Rape) - No. of Cases Reported": "Incest Rape",
    "Incest (Rape) No. of Victims - Upto 10 Years": "Incest <10 y.o.",
    "Incest (Rape) No. of Victims - (10-14) Years": "Incest 10-14 y.o.",
    "Incest (Rape) No. of Victims - (14-18) Years": "Incest 14-18 y.o.",
    "Incest (Rape) No. of Victims - (18-30)Years": "Incest 18-30 y.o.",
    "Incest (Rape) No. of Victims - (30-50) Years": "Incest 30-50 y.o.",
    "Incest (Rape) No. of Victims - Above 50 Years": "Incest >50 y.o.",
    "Incest (Rape) No. of Victims - Total Victims": "Incest Total Victims",
    "Other (Rape) - No. of Cases Reported": "Other Rape",
    "Other (Rape) No. of Victims - Upto 10 Years": "Other <10 y.o.",
    "Other (Rape) No. of Victims - (10-14) Years": "Other 10-14 y.o.",
    "Other (Rape) No. of Victims - (14-18) Years": "Other 14-18 y.o.",
    "Other (Rape) No. of Victims - (18-30)Years": "Other 18-30 y.o.",
    "Other (Rape) No. of Victims - (30-50) Years": "Other 30-50 y.o.",
    "Other (Rape) No. of Victims - Above 50 Years": "Other >50 y.o.",
    "Other (Rape) No. of Victims - Total Victims": "Other Total Victims",
    "Rape Cases (Total) - No. of Cases Reported": "Total Cases",
    "Rape Cases (Total) No. of Victims - Upto 10 Years": "Total <10 y.o.",
    "Rape Cases (Total) No. of Victims - (10-14) Years": "Total 10-14 y.o.",
    "Rape Cases (Total) No. of Victims - (14-18) Years": "Total 14-18 y.o.",
    "Rape Cases (Total) No. of Victims - (18-30)Years": "Total 18-30 y.o.",
    "Rape Cases (Total) No. of Victims - (30-50) Years": "Total 30-50 y.o.",
    "Rape Cases (Total) No. of Victims - Above 50 Years": "Total >50 y.o.",
    "Rape Cases (Total) No. of Victims - Total Victims": "Total Victims"
})

print(df.columns)
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
print(df.columns)
```

```python
df['Year'] = df['Year'].astype(int)


numerical_columns = ['Incest Rape', 'Other Rape']
df[numerical_columns] = df[numerical_columns].astype(float)
```

```python
duplicates = df.duplicated().sum()
print(f'Number of duplicate rows: {duplicates}')

df.drop_duplicates(inplace=True)
```

```python
yearly_trend = df.groupby('Year')['Total Cases'].sum()

yearly_trend.plot(kind='line', title='Total Rape Cases Reported Over Years', xlabel='Year', ylabel='Number of Cases')
```

```python
statewise_cases = df.groupby('States/ UTs/Cities')['Total Cases'].sum().sort_values(ascending=False)

statewise_cases.head(10).plot(kind='bar', title='Top 10 States/UTs by Total Rape Cases Reported', xlabel='State/UT', ylabel='Number of Cases')
```

```python
age_columns = [
    'Total <10 y.o.', 'Total 10-14 y.o.', 'Total 14-18 y.o.',
       'Total 18-30 y.o.', 'Total 30-50 y.o.', 'Total >50 y.o.'
]

age_group_distribution = df[age_columns].sum()

age_group_distribution.plot(kind='bar', title='Distribution of Victims Across Age Groups', xlabel='Age Group', ylabel='Number of Victims')
plt.xticks(rotation=45)
plt.show()
```

```python
incest_cases = df['Incest Rape'].sum()
other_cases = df['Other Rape'].sum()

pd.Series({'Incest Cases': incest_cases, 'Other Rape Cases': other_cases}).plot(kind='pie', title='Incest vs. Other Rape Cases')
```
